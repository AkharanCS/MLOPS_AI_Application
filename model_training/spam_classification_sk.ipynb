```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score,classification_report
from sklearn.naive_bayes import MultinomialNB
import json
import pickle as pkl
import mlflow
```

```python
df_main = pd.read_csv("enron_spam_data.csv")
df_main.fillna("Hello",inplace=True)
df_main["Message"] = df_main["Message"].astype("str")
```

```python
df_X = df_main.drop(columns="Spam/Ham",axis=1)
df_y = df_main["Spam/Ham"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(df_X,df_y)
```

```python
print(X_train["Message"])
```

```python
embed = CountVectorizer(max_features=50000)
X_embed_train = embed.fit_transform(X_train["Message"])
```

```python
dict = embed.get_feature_names_out()
print(dict)
print(len(dict))
if "university" in dict:
    print("True")
```

```python
mlflow.set_tracking_uri("http://127.0.0.1:5000")
```

```python
alpha_values = [0.1, 0.5, 1.0, 2.0, 5.0]
```

```python
model = MultinomialNB()
model.fit(X_embed_train, y_train)
```

```python
X_embed_test = embed.transform(X_test["Message"])
y_pred_NB = model.predict(X_embed_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_NB))
print("\nClassification Report:\n", classification_report(y_test, y_pred_NB))
```

```python
email = "please dont forget to report for the meeting today"
sample_embed = embed.transform([email])
print(model.predict(sample_embed))
```

```python
with open('naive_bayes_model.pkl', 'wb') as model_file:
    pkl.dump(model, model_file)

# Save the vectorizer
with open('count_vectorizer.pkl', 'wb') as vec_file:
    pkl.dump(embed, vec_file)
```
